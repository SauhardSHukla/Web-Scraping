### Problem statement
From this link, your task is to extract specific information from the article titled 'Web scraping with Python and BeautifulSoup' on the opensource website. You are required to print all the h2 tags found within the content of the article.

### Output Format :
User account menu

Main navigation

What is web scraping, and why do I need it?

How do I scrape a website in Python?

Extracting the HTML

```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url='https://opensource.com/article/21/9/web-scraping-python-beautiful-soup'
response = requests.get(base_url)
opne_source=BeautifulSoup(response.text)
print(opne_source)
```

```python
h2_tags=opne_source.select('h2')
print(h2_tags)
```

```python

lenght=len(h2_tags)
# print(lenght)
i=0
j=1
while i < len(h2_tags):
        print(f'{j}) {h2_tags[i].text.strip()}')
        j+=1
        i+=1
```

```python
#Different Approach to do it 
import requests
from bs4 import BeautifulSoup
response=requests.get("https://opensource.com/article/21/9/web-scraping-python-beautiful-soup")
soup=BeautifulSoup(response.text)
for i in soup.findAll("h2"):
    print(i.text.strip())
```
