```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://ultimateqa.com/simple-html-elements-for-automation')
response_Data = BeautifulSoup(response.text)
print(response_Data)
```

### Getting the Values Related to the Button (login & Click Me)

```python
click_data=response_Data.findAll('a',{'class':'et_pb_button et_pb_promo_button'})[0]
login_Data=response_Data.findAll('a',{'class':'et_pb_button et_pb_promo_button'})[1]
```

### Getting Value related to the Table 

```python
data=response_Data.find('table',{'id':"htmlTableId"})
```

```python
for data in response_Data.find('table',{'id':"htmlTableId"}).findAll('tr')[1:]:
  for i in data.findAll('td'):
    print(i.text)
```

```python
raw_data_file=[]
for table_row_data in response_Data.find('table',{'id':"htmlTableId"}).findAll('tr')[1:]:
  # print(data.td.text)
  row_data =[]
  for table_data in table_row_data.findAll('td'):
    # print(table_data.text.strip(),end='|\t')
    row_data.append(table_data.text)
  print(row_data)
  raw_data_file.append(row_data)
print(raw_data_file)
  
```

```python
pd.DataFrame(raw_data_file,columns=['col-1','col-2','col-3'])
```

```python

# If we want to find the value of option is 'Audi' or not in the options.

select_values=response_Data.find('select')
# print(select_values)
for values_audi in select_values.findAll('option'):
  # print(values_audi)
  if values_audi.text =='Audi':
    print(f'So we have the value:{values_audi['value']}')
```
