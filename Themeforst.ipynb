```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://themeforest.net/'
main_page_url = requests.get(url)
main_page_respinse = BeautifulSoup(main_page_url.text)
```

```python
all_item_nav_item = main_page_respinse.find('div',{'class':"shared-global_header-categories_menu_component__menuItem"})
```

```python
main_data =[]

new_all_item_Strip = all_item_nav_item.find_all('a')[1:3]
# new_all_item_Strip
data_href=[]
for ele in new_all_item_Strip:
  if ele :
    main_data.append(url[:-1]+ele['href'])
```

```python
new_data_page =[]
div_data =[]
for pg_data in  main_data:
  new_req=  requests.get(pg_data)
  new_req_data = BeautifulSoup(new_req.text)
  new_dashboard = new_req_data.find_all('div',{'class':'shared-item_cards-card_component__root'})
  div_data+=new_dashboard
```

```python
a_data = []
for data in div_data:
    link_form = data.find('a', {'class': 'shared-item_cards-grid-image_card_component__itemLinkOverlay'})
    if link_form is None:
        new_link_dta = data.find('a', {'class': 'shared-item_cards-preview_image_component__imageLink'})
        if new_link_dta and new_link_dta.img:  
            a_data.append([new_link_dta['href'], new_link_dta.img.get('title', 'No title')])
    else:
        a_data.append([link_form.get('href', 'No href'), link_form.get('title', 'No title')])
```
